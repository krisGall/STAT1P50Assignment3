Write a code to learn a simple regression model to predict the ratio of violent crimes based on (i) percentage of unemployed people and (ii) percentage of people with a bachelorâ€™s degree or higher education. Then explain
the impact of each of these two factors on violent crimes by interpreting the
regression coefficients.

```python
import sklearn.linear_model
import pandas
import matplotlib.pyplot

# Loading regression model
regression_model = sklearn.linear_model.LinearRegression(fit_intercept = False)

# Loading crime data
crime_data = pandas.read_csv('crime.csv')

# Loading relevant crime data
percentage_violent_crime = crime_data[['ViolentCrimesPerPop']]
percentage_unemployed = crime_data[['PctUnemployed']]
percentage_BS_or_More = crime_data[['PctBSorMore']]

# Training regression model for unemployed
regression_model_unemployed_trained = regression_model.fit(percentage_violent_crime,percentage_unemployed)

# Predicting crime for unemployed
predicted_crime_unemployed = regression_model_unemployed_trained.predict(percentage_violent_crime)

matplotlib.pyplot.figure()
matplotlib.pyplot.scatter(percentage_violent_crime, percentage_unemployed, s = 8)
matplotlib.pyplot.scatter(percentage_violent_crime, predicted_crime_unemployed, s = 8)
matplotlib.pyplot.draw()
print("Unemployed Coefficent:", regression_model_unemployed_trained.coef_)

# Training regression model for BS or higher
regression_model_bs_or_more_trained = regression_model.fit(percentage_violent_crime,percentage_BS_or_More)

# Predicting crime for BS or higher
predicted_crime_bs_or_more = regression_model_bs_or_more_trained.predict(percentage_violent_crime)

matplotlib.pyplot.figure()
matplotlib.pyplot.scatter(percentage_violent_crime, percentage_BS_or_More, s = 8)
matplotlib.pyplot.scatter(percentage_violent_crime, predicted_crime_bs_or_more, s = 8)
matplotlib.pyplot.draw()
print("BS or More Coefficient:", regression_model_bs_or_more_trained.coef_)
```

The coefficient factors show that violent crime is more impacted by the amount of population unemployed roughly 30% more then the amount of population with Bachelor degrees or higher
